```python
import os
import shutil
import random
import re
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Define folder paths
input_folder = '/content/drive/My Drive/text_files'
initial_folder = '/content/initial_files'
preprocessed_files = '/content/preprocessed_files'


# Create a function for natural sorting
def natural_sort_key(s):
    return [int(text) if text.isdigit() else text.lower() for text in re.split(r'(\d+)', s)]

# Function to create directories and copy files
def setup_folders_and_copy():
    # Remove the initial folder if it exists
    try:
        shutil.rmtree(initial_folder)
    except FileNotFoundError:
        pass

    # Copy files to the initial folder
    shutil.copytree(input_folder, initial_folder)

setup_folders_and_copy()
os.chdir(initial_folder)
files = os.listdir(initial_folder)

# Function to preprocess files
def preprocess_files(count=0):
    # Remove the preprocessed folder if it exists
    try:
        shutil.rmtree(preprocessed_files)
    except FileNotFoundError:
        pass

    # Create the preprocessed folder
    os.makedirs(preprocessed_files)

    # Process each file
    for i in range(1, count + 1):
        file = f"file{i}.txt"
        path = os.path.join(preprocessed_files, file)

        # Create empty files
        with open(path, 'w') as file:
            pass

# Main execution
if __name__ == "__main__":
    # Set working directory to the initial folder
    os.chdir(initial_folder)

    # Get the list of files and sort them naturally
    files = os.listdir(initial_folder)
    files = sorted(files, key=natural_sort_key)

    # Display 5 selectedly selected files
    selected_files = random.sample(files, 5)
    selected_files = sorted(selected_files, key=natural_sort_key)
    print(selected_files)

    # Count the number of files
    count = len(files)
    document_id = {}
    for i in range(1,count+1):
        document_id[i] = files[i-1]
    print(document_id)

    # Perform setup and preprocessing

    preprocess_files(count)

tokenized_folder = '/content/tokenized_files'
try:
    shutil.rmtree(tokenized_folder)
except FileNotFoundError:
    pass
os.makedirs("/content/tokenized_files")
for i in range(1, count+1):
    file = f"file{i}.txt"
    path = os.path.join(tokenized_folder, file)
    with open(path, 'w') as file:
        pass
```

```python
import os
import pandas as pd

def read_file_content(file_path):
    with open(file_path, 'r') as file:
        return file.read()

def write_file_content(file_path, content):
    with open(file_path, 'w') as file:
        file.write(content)

def preprocess_files(initial_folder, preprocessed_files, files, selected_files):
    print("Before preprocessing:\n")

    for filename in files:
        initial_path = os.path.join(initial_folder, filename)
        with open(initial_path, 'r') as initial_file:
            content = initial_file.read()

            if filename in selected_files:
                print(f"{filename} :")
                print(content, end="\n\n")

            content = content.lower()

        preprocessed_path = os.path.join(preprocessed_files, filename)
        write_file_content(preprocessed_path, content)

    print("----------------------------------------------------------------")
    print("After preprocessing:\n")

    for filename in selected_files:
        preprocessed_path = os.path.join(preprocessed_files, filename)
        content = read_file_content(preprocessed_path)

        print(f"{filename} :")
        print(content, end="\n\n")

if __name__ == "__main__":
    initial_folder = '/content/initial_files'
    preprocessed_files = '/content/preprocessed_files'

    # Assuming you have 'files' and 'selected_files' defined previously
    preprocess_files(initial_folder, preprocessed_files, files, selected_files)

```

```python
import os
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

def read_file_content(file_path):
    with open(file_path, 'r') as file:
        return file.read()

def write_content_to_file(file_path, content):
    with open(file_path, 'w') as file:
        file.write(content)

def tokenize_and_preprocess(files, preprocessed_files, tokenized_folder, selected_files):
    print("Before preprocessing:\n")

    for filename in files:
        preprocessed_path = os.path.join(preprocessed_files, filename)
        tokenized_path = os.path.join(tokenized_folder, filename)

        content = read_file_content(preprocessed_path)

        if filename in selected_files:
            print(f"{filename} :")
            print(content, end="\n\n")

        tokens = word_tokenize(content)

        write_content_to_file(preprocessed_path, str(tokens))
        write_content_to_file(tokenized_path, str(tokens))

    print("----------------------------------------------------------------")
    print("After preprocessing:\n")

    for filename in selected_files:
        preprocessed_path = os.path.join(preprocessed_files, filename)

        with open(preprocessed_path, 'r') as file:
            content = file.read()
            print(f"{filename} :")
            print(content, end="\n\n")

if __name__ == "__main__":
    initial_folder = '/content/initial_files'
    preprocessed_files = '/content/preprocessed_files'
    # Assuming you have 'files', 'preprocessed_files', 'tokenized_folder', and 'selected_files' defined previously
    tokenize_and_preprocess(files, preprocessed_files, tokenized_folder, selected_files)
```

```python
 from nltk.corpus import stopwords
import ast

def read_file_content(file_path):
    with open(file_path, 'r') as file:
        return file.read()

def write_file_content(file_path, content):
    with open(file_path, 'w') as file:
        file.write(content)

def remove_stopwords(files, selected_files, preprocessed_files):
    nltk.download('stopwords')
    stop_words = set(stopwords.words('english'))

    print("Before preprocessing:\n")

    for filename in files:
        path = os.path.join(preprocessed_files, filename)
        content = read_file_content(path)

        if filename in selected_files:
            print(f"{filename} :")
            print(content, end="\n\n")

        tokens = ast.literal_eval(content)
        stripped = [token for token in tokens if token.lower() not in stop_words]

        write_file_content(path, str(stripped))

    print("----------------------------------------------------------------")
    print("After preprocessing:\n")

    for filename in selected_files:
        path = os.path.join(preprocessed_files, filename)
        content = read_file_content(path)

        print(f"{filename} :")
        print(content, end="\n\n")

if __name__ == "__main__":
    initial_folder = '/content/initial_files'
    preprocessed_files = '/content/preprocessed_files'

    # Assuming you have 'files' and 'selected_files' defined previously
    remove_stopwords(files, selected_files, preprocessed_files)
```

```python
def read_file_content(file_path):
    with open(file_path, 'r') as file:
        return file.read()

def write_file_content(file_path, content):
    with open(file_path, 'w') as file:
        file.write(content)

def preprocess_and_strip_whitespace(files, selected_files, preprocessed_files):
    print("Before preprocessing:\n")

    for filename in files:
        path = os.path.join(preprocessed_files, filename)
        content = read_file_content(path)

        if filename in selected_files:
            print(f"{filename} :")
            print(content, end="\n\n")

        tokens = ast.literal_eval(content)
        stripped = [token for token in tokens if token.lower().strip()]

        write_file_content(path, str(stripped))

    print("----------------------------------------------------------------")
    print("After preprocessing:\n")

    for filename in selected_files:
        path = os.path.join(preprocessed_files, filename)
        content = read_file_content(path)

        print(f"{filename} :")
        print(content, end="\n\n")

if __name__ == "__main__":
    initial_folder = '/content/initial_files'
    preprocessed_files = '/content/preprocessed_files'

    # Assuming you have 'files' and 'selected_files' defined previously
    preprocess_and_strip_whitespace(files, selected_files, preprocessed_files)
```
